```python
x=np.arange(10)
plt.subplots_adjust(wspace=0.3,hspace=0.3)
plt.subplot(2,2,1)
plt.plot(x,-x**2,'-',color='g')
plt.subplot(2,2,2)
plt.plot(x,-x**3,'--',color='violet')
plt.subplot(2,2,3)
plt.plot(x,-2*x,':',color='r')
plt.subplot(2,2,4)
plt.plot(x,-2**x,'^',color='orange')
plt.show()
```
